# CRIC Convergence Plots

Takes the deviations saved by /kidney/convergence_plots/Generate and Save Deviations.ipynb and plots them

```python
import numpy as np
import pickle
import shap
import matplotlib.pyplot as pl

# runtime_kernel = pickle.load(open( "ke_runtime_dict.p", "rb" ))
mean_kernel = pickle.load(open( "kernel_expl_abs_dev_single_ref_truth_mean_dict.p", "rb" ))
max_kernel = pickle.load(open( "kernel_expl_abs_dev_single_ref_truth_max_dict.p", "rb" ))
mean_sampling = pickle.load(open( "sample_expl_abs_dev_single_ref_truth_mean_dict.p", "rb" ))
max_sampling = pickle.load(open( "sample_expl_abs_dev_single_ref_truth_max_dict.p", "rb" ))
```

```python
f = pl.figure(figsize=(6.5,4))
nsamples = [pow(10, v) for v in [3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5]]
#pl.axhline(0.063254 / 10, linestyle="--", color="#cccccc")
pl.axhline(0.063254 / 100, linestyle="--", color="#cccccc")
# pl.loglog(nsamples, max_kernel["aic"])
# pl.loglog(nsamples, max_kernel["bic"])
pl.loglog(nsamples, max_kernel[0.0], color=shap.plots.colors.old_blue_rgb, linewidth=3, label="Kernel SHAP")

pl.loglog(nsamples, max_sampling[10], color=shap.plots.colors.old_red_rgb, linewidth=3, label="IME (min 10)")
pl.loglog(nsamples, max_sampling[50], color=shap.plots.colors.old_red_rgb, linestyle="--", linewidth=3, label="IME (min 50)")
pl.loglog(nsamples, max_sampling["ns"], color=shap.plots.colors.old_red_rgb, linestyle="-.", linewidth=3, label="IME (uniform)")
pl.xlabel("Number of samples", fontsize=13)
pl.ylabel("Average absolute feature deviation\n(maximum over all features)", fontsize=13)
pl.gca().xaxis.set_ticks_position('bottom')
pl.gca().yaxis.set_ticks_position('left')
pl.gca().spines['right'].set_visible(False)
pl.gca().spines['top'].set_visible(False)
pl.legend(fontsize=13)
pl.savefig("max_kidney_agnostic_convergence.pdf")
pl.show()
```
